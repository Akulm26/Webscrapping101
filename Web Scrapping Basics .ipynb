## Scraping and Downloading data from [data.gov](https://www.data.gov)

```python
#Viewing csv file from the web directly

import csv, urllib.request

url = 'https://data.cdc.gov/api/views/hc4f-j6nb/rows.csv'
response = urllib.request.urlopen(url)
lines = [l.decode('utf-8') for l in response.readlines()]
cr = csv.reader(lines)

for row in cr:
    print(row)
```

```python
#Downloading csv file from the web onto your machine without pressing the "download" button. 

import requests

url = "https://data.cdc.gov/api/views/hc4f-j6nb/rows.csv"
#just a random link of a dummy file

r = requests.get(url)
#retrieving data from the URL using get method

with open("dummy.csv", 'wb') as f:
#giving a name and saving it in any required format
#opening the file in write mode

    f.write(r.content) 
#writes the URL contents from the server


```
